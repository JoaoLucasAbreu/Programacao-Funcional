## Criar Tabelas

```python
import sqlite3

conn = sqlite3.connect('p1.db')

cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS Course(
    courseId INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    studyArea TEXT NOT NULL
)
""")   

cursor.execute("""
CREATE TABLE IF NOT EXISTS Student(
    studentId INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    age INTEGER NOT NULL,
    cpf TEXT NOT NULL,
    courseId INTEGER,
    created DATETIME NOT NULL,
    FOREIGN KEY(courseId) REFERENCES Course(courseId)
)
""")

conn.commit()
conn.close()
```

```python
import sqlite3
import time
from datetime import date
from Student import Student
from Course import Course
from IPython.display import clear_output

conn = sqlite3.connect('p1.db')
cursor = conn.cursor()
cursor2 = conn.cursor()
minimo = cursor.execute('SELECT min(courseId)  FROM Course')
minimo = cursor.fetchone()
minimo = minimo[0]
maximo = cursor.execute('SELECT max(courseId)  FROM Course')
maximo = cursor.fetchone()
maximo = maximo[0]

passar1 = False
while passar1 == False:
    clear_output(wait = True) 
    print("Matrícula de Aluno\n" +
          "-----------------------")
    print("Opções:\n" +
          "[1] Adicionar aluno\n" +
          "[2] Remover aluno\n" +
          "[3] Listar alunos\n"+
          "[4] Atualizar matricula\n" +
          "[5] Sair\n")

    x = int(input("R:"))
    clear_output(wait=True) 

    if x == 1:
      print("[1] Adicionar aluno")
      student = Student(input("Nome do estudante:"), input("Idade:"), input("CPF:"))
      print("Cursos disponíveis\n" +
          "-----------------------")

      cursor.execute('SELECT * FROM Course')
      for linha in cursor.fetchall():
          print(f"[{linha[0]}] Curso: {linha[1]} - Area: {linha[2]}")

      passar2 = False
      while passar2 == False:
        cursoid = int(input("Curso de Interesse (id): "))
        if cursoid > maximo or cursoid < minimo:
          print("opção inválida\n")
          passar2 = False

        else:
          cursor.execute('INSERT INTO Student (name, age, cpf, courseId, created) VALUES (?, ?, ?, ?, ?)', (student.name, student.age, student.cpf, cursoid, date.today(),))
          conn.commit()
          print('cadastro gerado com sucesso')
          passar2 = True    
 
    elif x == 2:
      print("[2] Remover cadastro de aluno (por cpf)\n" +
          "-----------------------")
      cpf = input("CPF: ")
      cursor.execute('DELETE FROM Student WHERE cpf = ?', (cpf,))
      conn.commit()
      print('deletado com sucesso')
           
    elif x == 3: 
      print("[3] Listar alunos cadastrados\n" +
          "-----------------------")
      cursor.execute('SELECT * FROM Student')
      for linha in cursor.fetchall():
        print(f"Nome: {linha[1]}, idade: {linha[2]}, cpf: {linha[3]}")
        c = cursor2.execute('SELECT name FROM Course where courseId = ?', (linha[4],))
        c = cursor2.fetchone()
        c = c[0]
        print(f"Curso matriculado: {c}\n")


      print("alunos listados")
      time.sleep(10)

    elif x == 4:
      print("[4] Atualizar cadastrado de aluno (por cpf)\n" +
          "-----------------------")
      cpf = input("CPF: ")
      passar2 = False
      while passar2 == False:
        print("O que atualizar:\n" +
            "[1] Nome\n" +
            "[2] Idade\n" +
            "[3] Cpf\n"+
            "[4] Trocar curso escolhido\n" +
            "[5] Voltar")
        y = int(input("R:"))

        if y == 1:
          cursor.execute('UPDATE Student SET name = ? WHERE cpf = ?', (input("Novo nome:"), cpf,))
          print('update efetuado com sucesso')
          conn.commit()

        elif y == 2:
          cursor.execute('UPDATE Student SET age = ? WHERE cpf = ?', (input("Nova idade:"), cpf,))
          print('update efetuado com sucesso')
          conn.commit()

        elif y == 3:
          cursor.execute('UPDATE Student SET cpf = ? WHERE cpf = ?', (input("Novo cpf:"), cpf,))
          print('update efetuado com sucesso')
          conn.commit()

        elif y == 4:
          cursor.execute('SELECT * FROM Course')
          for linha in cursor.fetchall():
              print(f"[{linha[0]}] Curso: {linha[1]} - Area: {linha[2]}")

          passar3 = False
          while passar3 == False:
            cursoid = int(input("Novo curso de Interesse (id): "))
            if cursoid > maximo or cursoid < minimo:
              print("opção inválida\n")
              passar3 = False

            else:
              cursor.execute('UPDATE Student SET courseId = ? WHERE cpf = ?', (int(input("Novo curso de Interesse (id): ")), cpf,))
              passar3 = True
              conn.commit()
              print('update efetuado com sucesso')

        elif y == 5:
          passar2 = True

        else:
          print("Opção inválida")

    elif x == 5:
      passar1 = True   

    else:
      print("Opção inválida")

print("Programa Finalizado")
conn.close()
```
